# Python Project: Analysis of Employee Dataset

# Step 1: Import Necessary Libraries and load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
dataset = pd.read_csv('myexcel - myexcel.csv.csv')
dataset
```

# Preprocessing: Replace height with random numbers between 150 and 180

```python
dataset['Height'] = np.random.randint(150, 180, size=len(dataset))
dataset
```

# Analysis Task 1: Distribution of Employees Across Teams

```python
team_counts = dataset['Team'].value_counts()
print(team_counts)
team_counts.plot(kind='pie', autopct='%1.1f%%',textprops={'fontsize': 8})
plt.title('Distribution of Employees Across Teams')
plt.show()
```

# Analysis Task 2: Segregation of Employees by Position

```python
position_counts = dataset['Position'].value_counts()
print(position_counts)
position_counts.plot(kind='bar')
plt.title('Segregation of Employees by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

# Analysis Task 3: Predominant Age Group

```python
dataset['age_group'] = pd.cut(dataset['Age'], bins=[20, 30, 40, 50, 60, 100], labels=['20-29', '30-39', '40-49', '50-59', '60+'])
age_group_counts = dataset['age_group'].value_counts()

plt.figure(figsize=(8, 6))
plt.hist(dataset['Age'], bins=[20, 30, 40, 50, 60, 70], edgecolor='black')
plt.xticks([20, 30, 40, 50, 60, 70])
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

# Analysis Task 4: Team and Position with Highest Salary Expenditure

```python
team_salary = dataset.groupby('Team')['Salary'].sum()
position_salary = dataset.groupby('Position')['Salary'].sum()

print(team_salary)
print(position_salary)

team_salary.plot(kind='bar')
plt.title('Team Salary Expenditure')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.show()

position_salary.plot(kind='bar')
plt.title('Position Salary Expenditure')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

# Analysis Task 5: Correlation Between Age and Salary

```python
plt.scatter(dataset['Age'], dataset['Salary'])
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
print(dataset['Age'].corr(dataset['Salary']))
```

# Data Story

```python
print('The dataset contains', len(dataset), 'Employees.')
print('The average salary is', dataset['Salary'].mean())
print('The predominant age group is', dataset['age_group'].value_counts().idxmax())
print('The team with the highest salary expenditure is', team_salary.idxmax())
print('The position with the highest salary expenditure is', position_salary.idxmax())
print('There is a', dataset['Age'].corr(dataset['Salary']), 'correlation between age and salary.')
```
